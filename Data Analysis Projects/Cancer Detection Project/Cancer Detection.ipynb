# Machine Learning

```python
from sklearn.model_selection import train_test_split
```

```python
# seperating in test and train
```

```python
train, test = train_test_split(result_df, test_size= 0.25, random_state= 0, stratify = result_df['has_cancer'])
```

```python
train.head()
```

```python
test.head()
```

```python
train_X = train[train.columns[:30]]
test_X = test[test.columns[:30]]

train_Y = train['has_cancer']
test_Y = test['has_cancer']
```

```python
train_X.head()
```

```python
train_Y.head()
```

```python
#guassian work
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
```

```python
print(prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(test_Y, prediction))
```

```python
#confusion matrix
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(test_Y, prediction)
print('confusion_matrix', confusion_matrix)
plt.figure()
plt.matshow(confusion_matrix, cmap='Pastel1')
for x in range(0,2):
    for y in range(0,2):
        plt.text(x,y,confusion_matrix[x,y])
        
plt.ylabel('Actual Lablel')
plt.xlabel('Predicted Label')
plt.show
        
```

```python
def conf_matrix(test_Y,Prediction):
    confusion_matrix = metrics.confusion_matrix(test_Y, prediction)
    print('confusion_matrix', confusion_matrix)
    plt.figure()
    plt.matshow(confusion_matrix, cmap='Pastel1')
    for x in range(0,2):
        for y in range(0,2):
            plt.text(x,y,confusion_matrix[x,y])
        
    plt.ylabel('Actual Lablel')
    plt.xlabel('Predicted Label')
    plt.show
    print('TN: ',confusion_matrix[0][0])
    print('FN: ',confusion_matrix[0][1])
    print('FP: ',confusion_matrix[1][0])
    print('TP: ',confusion_matrix[1][1])
    
```

```python
conf_matrix(test_Y, prediction)
```

```python
#Fitting the logistic Regression algoritham
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
```

```python
print(prediction)
```

```python
print(accuracy_score(test_Y, prediction))

conf_matrix(test_Y, prediction)
```

```python
#Fitting the Kneighbour algoritham
from sklearn.neighbors import KNeighborsClassifier

model =  KNeighborsClassifier(n_neighbors = 5) 
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
```

```python
print(prediction)
print(accuracy_score(test_Y, prediction))

conf_matrix(test_Y, prediction)
```

```python
#fitting svm (linear)
from sklearn.svm import SVC
model = SVC(kernel = 'linear', random_state = 0)
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
```

```python
print(prediction)
print(accuracy_score(test_Y, prediction))

conf_matrix(test_Y, prediction)
```

```python
#fitting K-SVM
from sklearn.svm import SVC
model = SVC(kernel = 'rbf', random_state = 0)
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
```

```python
print(prediction)
print(accuracy_score(test_Y, prediction))

conf_matrix(test_Y, prediction)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_trainS = sc.fit_transform(train_X)
X_testS = sc.fit_transform(test_X)
```

```python
model = SVC(kernel = 'rbf', random_state = 0)
model.fit(X_trainS, train_Y)
prediction = model.predict(X_testS)
```

```python
print(prediction)
print(accuracy_score(test_Y, prediction))

conf_matrix(test_Y, prediction)
```

```python
#Desicion tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
```

```python
print(prediction)
print(accuracy_score(test_Y, prediction))

conf_matrix(test_Y, prediction)
```

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier(n_estimators = 10, criterion='entropy',random_state = 0)
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
```

```python
print(prediction)
print(accuracy_score(test_Y, prediction))

conf_matrix(test_Y, prediction)
```
